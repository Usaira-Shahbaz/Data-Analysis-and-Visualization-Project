Task 2

```python
import pandas as pd
import matplotlib.pyplot as plt

data_frame = pd.read_csv("/content/garments_worker_productivity.csv")
avg_productivity_per_quarter = data_frame.groupby("quarter")["actual_productivity"].mean()
ordered_quarters = ["Quarter1", "Quarter2", "Quarter3", "Quarter4"]
avg_productivity_per_quarter = avg_productivity_per_quarter.reindex(ordered_quarters)

plt.figure(figsize=(8, 5))
plt.plot(avg_productivity_per_quarter.index, avg_productivity_per_quarter.values, marker='x')

plt.xlabel("Quarterly Period")
plt.ylabel("Average Worker Productivity")
plt.title("Trends in Worker Productivity Over Different Quarters")

plt.legend(["Productivity Trend"])
plt.show()
```

Task 3

```python
df["department"] = df["department"].str.strip()
group_dep = df.groupby("department")["actual_productivity"].mean()

plt.figure(figsize=(8, 5))

group_dep.plot(kind="bar", color=["green", "blue"])


plt.xlabel("Department")
plt.ylabel("Average Productivity")
plt.title("Department-Wise Average Productivity")


plt.show()
```

Task 4

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["over_time"], df["actual_productivity"],alpha=0.8, color='grey')

plt.xlabel("Over Time")
plt.ylabel("Efficiency")
plt.title("Scatter Plot of Efficiency vs. Over Time")


plt.show()
```

# If higher over time, It will leads to higher efficiency, then we would see an upward trend in the scatter plot.

```python
df["absenteeism_time_in_minutes"] = df["idle_time"] * df["idle_men"] * 60
plt.figure(figsize=(10, 5))

plt.hist(df["absenteeism_time_in_minutes"], bins=15, color= "green", edgecolor='black')
plt.xlabel("Absenteeism Time (minutes)")
plt.ylabel("Frequency")
plt.title("Distribution of Worker Absenteeism Time")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
